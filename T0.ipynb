```python
import pandas as pd
from pandas.io.json import json_normalize
import warnings
from tabulate import tabulate
warnings.filterwarnings("ignore")
```

Aqui sacamos todos los tweets que esta en ingles, y dejamos solamente los primeros 300 mil lineas para un mejor funcionamiento

```python
raw_tweets = pd.read_json(r'farmers-protest-tweets-2021-03-5.json', lines=True, nrows=300000)
raw_tweets = raw_tweets[raw_tweets['lang']=='en']
raw_tweets.head(10)


```

Aqui definimos el top 10 retweeted

```python
def TopRetweet(tweetList):
    """
    Returns the most retweeted tweet in a list of tweets
    """
    order_by_retweet_count = tweetList.sort_values(by=['retweetCount'], ascending=False)
    return order_by_retweet_count[["id", "renderedContent", "retweetCount"]][0:10]

TopRetweet(raw_tweets)
```

Top 10 users

```python
def TopUser(raw_tweets):
    new_df = raw_tweets[['id', "user"]]
    new_df["username"] = new_df["user"].apply(lambda x: x["username"])
    count = new_df.groupby(["username"])["username"].count().sort_values(ascending=False)
    result = pd.DataFrame([count.index, count.values]).T.rename(columns = {0: 'User', 1: 'number_of_tweets'})
    return result[0:10]
TopUser(raw_tweets)
```

Top 10 dias con mas tweets

```python
def TopDays(raw_tweets):
    only_date_df = raw_tweets[['id', 'date']]
    only_date_df['Date'] = pd.to_datetime(only_date_df["date"]).dt.date
    count = only_date_df.groupby('Date')['Date'].count().sort_values(ascending=False)
    result = pd.DataFrame([count.index, count.values]).T.rename(columns = {0: 'Date', 1: 'number_of_tweets'})
    return result[0:10]

TopDays(raw_tweets)
```

Top 10 hashtags

```python
import re
def TopHashtags(raw_tweets):
    hashtag_map = {}
    raw_tweets['hashtag'] = raw_tweets['renderedContent'].apply(lambda x: re.findall(r"#(\w+)", x))
    for row in raw_tweets['hashtag']:
        for item in row:
            if item in hashtag_map:
                hashtag_map[item] += 1
            else:
                hashtag_map[item] = 1
    sorted_map = sorted(hashtag_map.items(), key=lambda item: item[1], reverse=True)
    df = pd.DataFrame(sorted_map, columns=['hashtag', 'count'])
    return df[0:10]
TopHashtags(raw_tweets)
```

Funcion main

```python
def main():
    funtion_map = {1: TopRetweet, 2: TopUser, 3: TopDays, 4: TopHashtags}
    print("Twitter analysis")
    print("1. Top 10 Retweeted Tweets \n2. Top 10 Users \n3. Top 10 Days \n4. Top 10 Hashtags \n")
    user_input = input("inpunt an option:")
    if user_input in ['1', '2', '3', '4']:
        result = funtion_map[int(user_input)](raw_tweets)
        print("result:")
        print(tabulate(result, headers='keys', tablefmt='psql'))
    else:
        print("invalid input")
        return main()


```

```python
if __name__ == "__main__":
    main()
```
